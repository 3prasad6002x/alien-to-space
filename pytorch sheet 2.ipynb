```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/PacktWorkshops/The-Deep-Learning-with-PyTorch-Workshop/master/Chapter02/Exercise2.02/energydata_complete.csv")
```

```python
data=data.drop(columns=["date"])
```

```python
data
```

```python
columns=data.columns
column_numeric=data._get_numeric_data().columns
x=list(set(columns)-set(column_numeric))
x
```

```python
data.isnull().sum()
```

```python
outliers={}
for i in range(data.shape[1]):
    min_t=data[data.columns[i]].mean()-(3*data[data.columns[i]].std())
    max_t=data[data.columns[i]].mean()+(3*data[data.columns[i]].std())
    count=0
    for j in data[data.columns[i]]:
        if j<min_t or j>max_t:
            count+=1
    percentage=count/data.shape[0]
    outliers[data.columns[i]]="%.3f" % percentage
outliers    
```

```python
X=data.iloc[:,1:]
Y=data.iloc[:,0]
```

```python
X=(X-X.min())/(X.max()-X.min())
X
```

```python
from sklearn.model_selection import train_test_split
X_shuffle=X.sample(frac=1,random_state=0)
Y_shuffle=Y.sample(frac=1,random_state=0)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
dev_shape=X_test.shape[0]/X_train.shape[0]
X_train1,X_dev,Y_train1,Y_dev=train_test_split(X_train,Y_train,test_size=dev_shape,random_state=0)
```

```python
print(X_train1.shape,X_test.shape,X_dev.shape)
print(Y_train1.shape,Y_test.shape,Y_dev.shape)
dev_shape
```

```python
import torch
import torch.nn as nn
torch.manual_seed(0)
```

```python
X_train1=torch.tensor(X_train1.values).float()
X_test=torch.tensor(X_test.values).float()
X_dev=torch.tensor(X_dev.values).float()
Y_train1=torch.tensor(Y_train1.values).float()
Y_test=torch.tensor(Y_test.values).float()
Y_dev=torch.tensor(Y_dev.values).float()
```

```python
model=nn.Sequential(nn.Linear(X_train1.shape[1],100),
                    nn.ReLU(),
                   nn.Linear(100,50),
                   nn.ReLU(),
                   nn.Linear(50,25),
                   nn.ReLU(),
                   nn.Linear(25,1))
```

```python
loss_function=torch.nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
for i in range(1000):
    Y_pred=model(X_train1).squeeze()
    loss=loss_function(Y_pred,Y_train1)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if i%100==0:
        print(i,loss.item())
```

```python
pred=model(X_dev[0])
print("ground truth",Y_dev[0].item(), "predicton",pred.item())
```
