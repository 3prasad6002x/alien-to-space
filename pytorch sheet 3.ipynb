```python
data.isnull().sum()
data.describe(include="all")
```

```python
outliers={}
for i in range(data.shape[1]):
    min_t=data[data.columns[i]].mean()-(3*data[data.columns[i]].std())
    max_t=data[data.columns[i]].mean()+(3*data[data.columns[i]].std())
    count=0
    for j in data[data.columns[i]]:
        if j<min_t or j>max_t:
            count+=1
    percentage=count/data.shape[0]
    outliers[data.columns[i]]="%.3f"%percentage
```

```python
outliers
```

```python
X=data.iloc[:,1:]
Y=data.iloc[:,0]
```

```python
X=(X-X.max())/(X.max()-X.min())
```

```python
X
```

```python
X_shuffle=X.sample(frac=1,random_state=0)
Y_shuffle=Y.sample(frac=1,random_state=0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_shuffle,Y_shuffle,test_size=0.2,random_state=0)
dev_shape=X_test.shape[0]/X_train.shape[0]
X_train1,X_dev,Y_train1,Y_dev=train_test_split(X_train,Y_train,test_size=dev_shape,random_state=0)
```

```python
print(X_train1.shape,X_dev.shape,X_test.shape,Y_train1.shape,Y_dev.shape,Y_test.shape)
```

```python
import torch
import torch.nn as nn
```

```python
X_train1=torch.tensor(X_train1.values).float()
X_train1.to("cuda")
X_test=torch.tensor(X_test.values).float()
X_test.to("cuda")
X_dev=torch.tensor(X_dev.values).float()
X_dev.to("cuda")
Y_train=torch.tensor(Y_train1.values).float()
Y_train.to("cuda")
Y_test=torch.tensor(Y_test.values).float()
Y_test.to("cuda")
Y_dev=torch.tensor(Y_dev.values).float()
Y_dev.to("cuda")
```

```python
model=nn.Sequential(nn.Linear(X_train1.shape[1],1000),
                   nn.ReLU(),
                   nn.Linear(1000,500),
                   nn.ReLU(),
                   nn.Linear(500,250),
                   nn.ReLU(),
                   nn.Linear(250,125),\
                   nn.ReLU(),
                   nn.Linear(125,75),
                   nn.Linear(75,25),
                   nn.Linear(25,1))
model.to("cuda")
```

```python
loss_function=torch.nn.MSELoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
X_train1.to("cuda")
```

```python
for i in range(3000):
    Y_pred=model(X_train1.to("cuda")).squeeze()
    loss=loss_function(Y_pred,Y_train.to("cuda"))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if i%100==0:
        print(i,loss.item())
    
```

```python
pred = model(X_dev[0].to("cuda"))
print("Ground truth:", Y_dev[0].item(), "Prediction:", pred.item())
```
